```python
subMEDF['Location']='ME'
subCTDF['Location']='LIS'
subFLDF['Location']='FL'
```

```python
framesALL= [subMEDF, subCTDF, subFLDF]
mergeALL = pd.concat(framesALL)
mergeALL
```

```python
# just summer months
summer = mergeALL[(mergeALL["MM"] == 6) | (mergeALL["MM"] == 7) | (mergeALL['MM'] == 8) | (mergeALL["MM"] == 9)]
summer
```

```python
#Remove 999
SummerDat = summer[summer.WTMP != 999.0]
```

```python
#Rename column headers
SummerDat=SummerDat.rename(columns={'#YY': 'YY', 'MM': 'MM', 'DD': 'DD', 'hh': 'hh', 'mm': 'mm', 'WTMP': 'WTMP'})
```

```python
#Left off here ... need to make three subplots
#First all summer data for 10 years
#Second all summer data for 2018
#Collection month for year of 2018
```

```python
#All summer data for 10 years
sns.boxplot(x='Location',y='WTMP',data=SummerDat, order=['ME','LIS','FL'])
plt.ylabel('Sea Surface Temperature (°C)')
plt.xlabel('Buoy Site')
plt.title('')
```

```python
#Summer data for 2018 only
CollectionYear = SummerDat[SummerDat.YY == 2018]
```

```python
#Plot CollectionYear
sns.boxplot(x='Location',y='WTMP',data=CollectionYear, order=['ME','LIS','FL'])
plt.ylabel('Sea Surface Temperature (°C)')
plt.xlabel('Buoy Site')
plt.title('')
plt.show()
```

```python
#Need to make MMDD column
CollectionYear["MMDD"] = CollectionYear["MM"].astype(str) + CollectionYear["DD"].astype(str)
```

```python
CollectionYear
```

```python
#All summer data for 2018
sns.boxplot(x='Location',y='WTMP',data=CollectionYear, order=['ME','LIS','FL'])
plt.ylabel('Sea Surface Temperature (°C)')
plt.xlabel('Buoy Site')
plt.title('')
```

```python
#Need to make a dataframe For ME from 7/12-8/12
CollectionME = CollectionYear.loc[CollectionYear['MMDD'].isin(['712','812'])]
```

```python
#Now I want only the Maine location
CollectionMEonly = CollectionME.loc[CollectionME['Location'] == 'ME']
```

```python
#Dataframe for NY 6/3-7/3
CollectionNY = CollectionYear.loc[CollectionYear['MMDD'].isin(['63','73'])]
```

```python
CollectionNYonly = CollectionNY.loc[CollectionNY['Location'] == 'LIS']
```

```python
#Dataframe for FL 6/25-7/25
CollectionFL = CollectionYear.loc[CollectionYear['MMDD'].isin(['625','725'])]
```

```python
CollectionFLonly = CollectionFL.loc[CollectionFL['Location'] == 'FL']
```

```python
#Merge dataframes together to make one for plots of month before collection
CollectionMonthDF= [CollectionMEonly, CollectionNYonly, CollectionFLonly]
AllCollectionMonthDF = pd.concat(CollectionMonthDF)
```

```python
#Plot Collection Month
sns.boxplot(x='Location',y='WTMP',data=AllCollectionMonthDF, order=['ME','LIS','FL'])
plt.ylabel('Sea Surface Temperature (°C)')
plt.xlabel('Buoy Site')
plt.title('')
```

```python
my_pal = {"ME": "xkcd:azure", "LIS": "xkcd:lighter green", "FL":"xkcd:barbie pink"}
```

```python
#Now want to make one plot with three subplots all with the same axes

sns.set(font_scale = 2, style='whitegrid')
fig, axes = plt.subplots(2, 2, figsize=(15, 15), sharey=True)
fig.suptitle('Buoy Sea Surface Temperature Data at Different Timescales', y=1.1, fontsize = 25)

sns.boxplot(ax=axes[0,1], x='Location',y='WTMP',data=SummerDat, order=['ME','LIS','FL'], palette=my_pal)

sns.boxplot(ax=axes[1,0], x='Location',y='WTMP',data=CollectionYear, order=['ME','LIS','FL'], palette=my_pal)

sns.boxplot(ax=axes[1,1], x='Location',y='WTMP',data=AllCollectionMonthDF, order=['ME','LIS','FL'], palette=my_pal)

axes[0,1].set_ylabel('Sea Surface Temperature (°C)', fontsize = 20)
axes[1,0].set_ylabel('Sea Surface Temperature (°C)', fontsize = 20)
axes[1,1].set_ylabel('Sea Surface Temperature (°C)', fontsize = 20)

axes[0,1].set_xlabel('Buoy Sites', fontsize = 20)
axes[1,0].set_xlabel('Buoy Sites', fontsize = 20)
axes[1,1].set_xlabel('Buoy Sites', fontsize = 20)

axes[0,1].set_title('B )   Decade',fontsize = 25,loc = 'left')
axes[1,0].set_title('C )   Summer', fontsize = 25, loc = 'left')
axes[1,1].set_title('D )   30 Days', fontsize = 25, loc = 'left')

fig.tight_layout()
```
